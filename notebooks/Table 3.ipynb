# Table 3

Synthetic nonlinear experiment, second sample-size setting. Run all cells to generate the results and display Table 3.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
while not (ROOT / "src" / "puate").exists():
    if ROOT.parent == ROOT:
        raise RuntimeError("Could not locate the repository root.")
    ROOT = ROOT.parent

SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

import pandas as pd
import warnings

warnings.filterwarnings("ignore")

from puate.paper_reproduction import run_table_3_experiment
from puate.reporting import summarize_trials, combine_summaries

pd.set_option("display.max_columns", None)
```

## Configuration

Set `FORCE_RERUN = True` to recompute the experiment and overwrite the saved outputs under `results/table_3`.

```python
FORCE_RERUN = False
BASE_SEED = 0
N_TRIALS = 5000
OUTPUT_ROOT = ROOT / "results" / "table_3"
```

```python
results = run_table_3_experiment(
    output_root=OUTPUT_ROOT,
    force=FORCE_RERUN,
    n_trials=N_TRIALS,
    base_seed=BASE_SEED,
    verbose=True,
)

censoring = results["censoring"]
case_control = results["case_control"]
```

```python
censoring_summary = summarize_trials(
    censoring.estimates,
    censoring.variances,
    true_ate=censoring.true_ate,
    ci_sample_size=censoring.ci_sample_size,
)

case_control_summary = summarize_trials(
    case_control.estimates,
    case_control.variances,
    true_ate=case_control.true_ate,
    ci_sample_size=case_control.ci_sample_size,
)

paper_table = combine_summaries(censoring_summary, case_control_summary)
paper_table.round(2)
```

```python
print(paper_table.round(2).to_latex(float_format="%.2f"))
```
