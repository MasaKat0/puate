# Figure 5

Density plot for the IHDP response surface A experiment. Run all cells to generate the results and display Figure 5.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
while not (ROOT / "src" / "puate").exists():
    if ROOT.parent == ROOT:
        raise RuntimeError("Could not locate the repository root.")
    ROOT = ROOT.parent

SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

import pandas as pd
import warnings

warnings.filterwarnings("ignore")

from puate.paper_reproduction import run_table_4_experiment
from puate.reporting import plot_two_setting_density

pd.set_option("display.max_columns", None)
```

## Configuration

This notebook reads experiment outputs under `results/table_4`. Set `FORCE_RERUN = True` to recompute them before plotting.

```python
FORCE_RERUN = False
BASE_SEED = 0
N_TRIALS = 1000
OUTPUT_ROOT = ROOT / "results" / "table_4"
```

```python
results = run_table_4_experiment(
    output_root=OUTPUT_ROOT,
    force=FORCE_RERUN,
    n_trials=N_TRIALS,
    base_seed=BASE_SEED,
    verbose=True,
)

censoring = results["censoring"]
case_control = results["case_control"]
```

```python
fig = plot_two_setting_density(
    censoring.estimates,
    case_control.estimates,
    true_ate=censoring.true_ate,
)
fig
```
